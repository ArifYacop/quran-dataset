```python
try: 
    cur.execute("CREATE TABLE IF NOT EXISTS arabic (surah int ,ayah int, text varchar);")
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
```

```python
surah = 1
```

```python
while surah <= 114:
    url = ("https://raw.githubusercontent.com/risan/quran-json/main/dist/chapters/{}.json".format(surah))
    response = urlopen(url)
    data_json = json.loads(response.read())
    try: 
        for verse in data_json['verses']:
            id = verse['id']
            text = verse['text']
            cur.execute("INSERT INTO arabic (surah, ayah, text) \
                     VALUES (%s, %s, %s)", \
                     ("{}".format(surah),id,text))
    except psycopg2.Error as e:
        print("Error: Inserting Rows")
        print (e)
    print('Complete surah number: ',surah)
    surah += 1
```

```python
try: 
    cur.execute("SELECT * FROM arabic;")
except psycopg2.Error as e: 
    print("Error: select *")
    print (e)

row = cur.fetchone()
while row:
   print(row)
   row = cur.fetchone()
```

```python
try: 
    cur.execute("select count(*) from arabic")
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
    
print(cur.fetchall())
```

```python
try: 
    cur.execute("CREATE TABLE IF NOT EXISTS indotranslater (language_id int,surah int ,ayah int ,transliteration varchar ,meaning varchar);")
except psycopg2.Error as e: 
    print("Error: Issue creating table")
    print (e)
```

```python
surah = 1
while surah <= 114:
    url = ("https://raw.githubusercontent.com/risan/quran-json/main/dist/chapters/id/{}.json".format(surah))
    response = urlopen(url)
    data_json = json.loads(response.read())
    try: 
        for verse in data_json['verses']:
            id = verse['id']
            transliteration = verse['transliteration']
            translation = verse['translation']            
            cur.execute("INSERT INTO indotranslater (language_id,surah,ayah,transliteration,meaning) \
                     VALUES (%s, %s, %s, %s, %s)", \
                     (2,"{}".format(surah), id, transliteration, translation))
    except psycopg2.Error as e:
        print("Error: Inserting Rows")
        print (e)
    print('Complete surah number: ',surah)
    surah += 1
```
